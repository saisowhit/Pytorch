```python
import json
import pandas as pd
import seaborn as sns
from pygments import highlight
from tqdm impor tqdm
import matplotlib.pyplot as plt
from google_play_scraper import Sort,reviews,app
```

```python
app_packages=['com.an']
```

```python
app_infos=[]
for ap in tqdm(app_packages):
    info=app(ap,language='em',country='US')
    del info['comments']
    app_infos.append(info)
    
```

```python
def print_json(json_object):
    json_str=json.dumps(json_object,indent=2,sort_keys=True,default=str)
print(highlight(json_str,JsonLexer(),TerminalFormatter()))
```

```python
print_json(app_infos[0])
```

```python
## Get App review
```

```python
app_reviews=[]
for ap in tqdm(app_packages):
    for score in range(1,6):
        for sort_order in[Sort.MOST_RELEVANT,Sort.NEWEST]:
            rvs=reviews(ap,language='en',country='us',sort=sort_order,count=200 if score==3 else 100)
            for r in rvs:
                r['sortOrder']='most_relevant' if sort_order==Sort.MOST_RELEVANT else 'newest'
                r['appId']=ap
            app_reviews.extend(rvs)
```

```python
print_json(app_reviews[0])
```
